```python
import numpy as np
import pandas as pd
```

```python
def ode(a, t):
    return a * t + 7

def ode2(a, t):
    return a*t**2 - 16*t + 7
```

```python
np.random.seed(11)
t = np.arange(0, 300)
a = np.random.randn(10000)

dy_dt = [] 
for i in range(10000):
    dy_dt.append(ode2(a[i], t))
```

```python
df1 = pd.DataFrame(np.array(dy_dt))
df1["a"] = a
df1.to_excel("simulated_data2.xlsx", index=False)
df1.head()
```
